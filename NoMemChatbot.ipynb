```python
# Recommend for demos -> fast and economial model is gemini 2.0 flash
model_name = "gemini-2.0-flash"

# llm instance
llm = ChatGoogleGenerativeAI(model=model_name,
                             convert_system_message_to_human =True,#to mkae it predictable with langchain
                             temperature = 0.7 )#to encorage creativity and responses
print(f"Ready with model: {model_name}")
```
